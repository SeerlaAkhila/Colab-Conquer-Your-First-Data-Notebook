```python
# ===============================
# Iris Dataset EDA - Complete Solution
# ===============================

# Step 1: Import Required Libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load the Dataset
url = "https://bit.ly/4nejNue"
df = pd.read_csv(url)

# Step 3: Display the First Few Rows
print("ðŸ”¹ First 5 Rows of Dataset:\n")
print(df.head())

# Step 4: Dataset Info
print("\nðŸ”¹ Dataset Information:\n")
df.info()

# Step 5: Show Basic Statistics
print("\nðŸ”¹ Mean of Numerical Features:\n")
print(df.mean(numeric_only=True))

print("\nðŸ”¹ Median of Numerical Features:\n")
print(df.median(numeric_only=True))

print("\nðŸ”¹ Standard Deviation of Numerical Features:\n")
print(df.std(numeric_only=True))

print("\nðŸ”¹ Full Summary Statistics:\n")
print(df.describe())

# Step 6: Visualize - Histograms of All Numerical Features
print("\nðŸ”¹ Generating Histograms...\n")
df.hist(figsize=(10, 8), bins=15, color='skyblue', edgecolor='black')
plt.suptitle("Histograms of All Features", fontsize=16)
plt.show()
```
